Con los datos mostrados en la tabla y haciendo uso de la librería Numpy, resuelve el ejercicio utilizando bucles anidados:

◼ Muestra por pantalla cuántos alumnos suspendieron cada asignatura. Ejemplo: Inglés se ha suspendido por 3 alumnos

◼ Realiza la media de las notas de cada alumno y muéstralas por pantalla. Ejemplo: El alumno1 ha tenido una nota media de 7.25

◼ Muestra por pantalla los nombres de los alumnos que obtuvieron

aprobado el curso son: Alumno1, ….

```python
#Importación de las librarias
import pandas as pd
import numpy as np
```

```python
#df para visualización de las notas
boletin = {'Latim': [8, 7, 10, 4, 9], 'Castellano': [8, 6, 7, 4, 8], 'Francés': [9, 7, 8, 3, 9], 'Inglés': [4, 2, 9, 2, 6]}
boletin_df = pd.DataFrame(boletin, index=['Alumno 1', 'Alumno 2', 'Alumno 3', 'Alumno 4', 'Alumno 5'])
display(boletin_df)
#---------------------------------------------------------------------------------------------

#matriz numpy para manipulación de los dados
boletin_np = np.array([[8, 8, 9, 4], [7, 6, 7, 2], [10, 7, 8, 9], [4, 4, 3, 2], [9, 8, 9 ,6]])

#Alumnos suspensos por asignaturas
asignaturas = np.array(['Latim', 'Castellano', 'Francés', 'Inglés'])
lista2 = np.array([],dtype=int)

for indice, nota in enumerate(boletin_np.T):
    for nota_baja in nota:
        if nota_baja < 5:
            lista2 = np.append(lista2,nota_baja)
            
    print(f'{asignaturas[indice]} se ha suspendido por {len(lista2)} alumnos')
    lista2 = np.delete(lista2,0,0)
print()

#haciendo el promedio de los estudiantes
alumnos = ['Alumno 1', 'Alumno 2', 'Alumno 3', 'Alumno 4', 'Alumno 5']
aprovados  = []

for indice, notas_alumnos in enumerate(boletin_np):
    soma_notas=0
    for n in notas_alumnos:
        soma_notas+=n
        media = soma_notas/len(notas_alumnos)
    print(f'El {alumnos[indice]} ha tenido una nota media de {soma_notas/len(notas_alumnos)}')
    if media > 5:
        aprovados.append([alumnos[indice], media])
        
#Imprimindo la lista de alunmos aprovados         
print(f'\nOs alumnos aprovados são: ', end=' ')
for ind, item in enumerate(aprovados):
    print(item[0], end=',')
```

```python
import numpy as np

#creando un array de cuatro dimensiones
array_num_np = np.array(range(1,25)).reshape(4,6)
print(array_num_np, '\n')

#partirmos el  array (lista_num_np) en  trozos.
array1, array2 = np.vsplit(array_num_np,2)

print(f'1º Trozo\n{array1}')
print(f'\n2º Trozo\n{array2}')

#sumamos los elementos del array en función de los últimos ejes
suma_np = array1+array2
print('\nSuma')
print(suma_np)
```

5 . Utilizando el archivo csv: COVID_01-01-2021, crea un programa que cree un gráfico de líneas en el que se muestre el total de casos confirmados, el total de casos fallecidos y el total de pacientes recuperados, y que compare sólo los países cuyo número de fallecidos sea superior a 150.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#cogemos los casos de los fallecidos que son superiores a 150 y sumamos todos los casos
#(confirmados, Fallecidos, Recuperados)
total_df = enero_df[['Country_Region','Confirmed', 'Deaths', 'Recovered']]
total_df = total_df.loc[total_df['Deaths']>150].groupby('Country_Region').sum()

#el gráfico de lineas enseñando los datos de (confirmados, Fallecidos, Recuperados)
#para este caso yo elegí hacer en el mismo gráfico, porque se pode comparar lo datos
#pero se puede hacer cada uno en un gráfico por separado.
display(total_df)
sns.set_style('whitegrid')
#como son muchos países para enseñar en el gráfico, hice un "head" con los 25 primeros países de la lista para plotar.
sns.lineplot(data = total_df.head(25)).figure.set_size_inches(18,5)
#ajuste la rotación de la label do eje "X" con "plt.xticks"
plt.xticks(rotation=45)
plt.show()
```

6 - Crea un programa que visualice los casos de coronavirus confirmados en el mundo durante el primer trimestre de 2021 y muéstralo en un gráfico, que incluya un título.




```python
#creé un dataframe para cada mes de primer trimestre
enero_df = pd.read_csv(r'COVID_01-01-2021.csv')
febrero_df = pd.read_csv(r'COVID_01-02-2021.csv')
marzo_df = pd.read_csv(r'COVID_01-03-2021.csv')

#concatené los tres dataframes
trimestre_df = pd.concat([enero_df, febrero_df, marzo_df])

#usé el groupby para agrupar los casos por países 
#usé (as_index=False) porque el índice estaba como países y en el gráfico me salía un error 
conf_df = trimestre_df.loc[:,['Country_Region', 'Confirmed']].groupby('Country_Region', as_index=False).sum()
display(conf_df)

#ajuste el color de fundo do gráfico
sns.set_style('darkgrid')

#plot en el gráfico los 20 primeros países de DataFrame 
conf_graf = sns.barplot(data= conf_df.head(20), x='Country_Region',y= 'Confirmed')

#Incluí un título 
conf_graf.set_title('Casos de coronavirus confirmados en el mundo durante el primer trimestre de 2021')

#ajuste el tamaño del gráfico para que se puede visualizar mejor los datos
conf_graf.figure.set_size_inches(15, 6)
plt.xticks(rotation=45)
plt.show()
```

7 - Crea un programa que muestre la evolución de los casos confirmados y de pacientes recuperados de coronavirus, en las diferentes provincias de España, durante el primer trimestre de 2021. Para este último apartado también se aceptaría que mostréis en un único gráfico la evolución tanto de los casos confirmados como de los pacientes recuperados de coronavirus en España, durante el primer trimestre de 2021


```python
#creando un dataframe con los casos confirmados y de pacientes recuperados en el España durante el primer trimestre
spain_df = trimestre_df.loc[trimestre_df['Country_Region']=='Spain', ['Province_State' ,'Country_Region','Confirmed', 'Recovered', 'Last_Update']]
display(spain_df)
```

```python
#el plot con los casos recuperados
sns.set_style('darkgrid')
sns.barplot(data=spain_df, x= 'Province_State', y= 'Recovered').figure.set_size_inches(15,6)
plt.xticks(rotation=45)
plt.show()

#el plot con los casos confirmados
sns.set_style('darkgrid')
sns.barplot(data=spain_df, x= 'Province_State', y= 'Confirmed').figure.set_size_inches(15,6)
plt.xticks(rotation=45)
plt.show()
```

```python
#aqui un único gráfico con la evolución tanto de los casos confirmados como de los pacientes recuperados de coronavirus en España, durante el primer trimestre de 2021
#hice un dataframe con ('Province_State', 'Confirmed', 'Recovered')
total_spain_df = spain_df[['Province_State', 'Confirmed', 'Recovered']]

#ajuste el índece para 'Province_State' para se visualize los nombres en le gráfico
total_spain_df= total_spain_df.set_index('Province_State')
display(total_spain_df.head(15))

#plot los dos casos en el mismo gráfico para que se pueda visualizar mejor la evolución
sns.relplot(data=total_spain_df.head(25)).figure.set_size_inches(12,4)
plt.xticks(rotation=45)
plt.show()
```
